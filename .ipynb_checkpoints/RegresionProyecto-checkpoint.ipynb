```python
###MATRIZ DE DEFINICION DE CLIENTES POR RANGO DE EDAD
# Create a function to define group age
def get_groupage(customer):
    age = customer
    
    if age < 12:
        return 'NIÃ‘O'
    elif age < 20:
        return 'ADOLECENTE'
    elif age < 40:
        return 'JOVEN'
    elif age < 65:
        return 'ADULTO'    
    else:
        return 'OTROS'
    
credit['person'] = credit['AGE'].apply(get_groupage)
```

```python
# CLIENTES POR ESTADO DE MOROSIDAD, GRUPO AL QUE PERTENECE, ESTADO CIVIL
cp = sns.catplot('MARRIAGE', col="IN_DEFAULT", data=credit, kind='count', hue='person', order=[0,1,2,3,4], 
               hue_order=['NIÃ‘O','ADOLECENTE','JOVEN','ADULTO','OTROS'], aspect=1)
cp.fig.suptitle('CLIENTES POR ESTADO DE MOROSIDAD, GRUPO AL QUE PERTENECE, ESTADO CIVIL',y=1.03)
cp.set_xlabels('ESTADO CIVIL')
cp.set_xticklabels(["OTRO", "CASADO", "SOLTERO", "DIVORCIADO"])
```

```python
# CLIENTES POR ESTADO DE MOROSIDAD, GRUPO AL QUE PERTENECE Y GENERO
cp = sns.catplot('SEX', col="IN_DEFAULT", data=credit, kind='count', hue='person', order=[1,2], 
               hue_order=['NIÃ‘O','ADOLECENTE','JOVEN','ADULTO','OTROS'], aspect=1)
cp.fig.suptitle('CLIENTES POR ESTADO DE MOROSIDAD, GRUPO AL QUE PERTENECE Y GENERO',y=1.03)
cp.set_xlabels('GENERO')
cp.set_xticklabels(["MALE", "FAME"])
```

```python
##CREAMOS UN GRUPO DEL LIMITE DE CREDITO
# Create a function to define group age
def get_creditcat(customer):
    limitbal = customer
    
    if limitbal < 100000:
        return 'class10'
    elif limitbal < 200000:
        return 'class9'
    elif limitbal < 300000:
        return 'class8'
    elif limitbal < 400000:
        return 'class7'
    elif limitbal < 500000:
        return 'class6'
    elif limitbal < 600000:
        return 'class5'
    elif limitbal < 700000:
        return 'class4'
    elif limitbal < 800000:
        return 'class3'
    elif limitbal < 900000:
        return 'class2'    
    else:
        return 'class1'
```

```python
credit['creditclass'] = credit['LIMIT_BAL'].apply(get_creditcat)
# CANTIDAD DE CLIENTES POR ESTADO CIVIL, CLASE DE CREDITO Y ESTADO CIVIL
cp = sns.catplot('MARRIAGE', col="IN_DEFAULT", data=credit, kind='count', hue='creditclass', order=[0,1,2,3,4], 
               hue_order=['class1','class2','class3','class4','class5','class6','class7','class8','class9','class10'],
               aspect=1)
cp.fig.suptitle('CANTIDAD DE CLIENTES POR ESTADO CIVIL, CLASE DE CREDITO Y ESTADO CIVIL', y=1.03)
cp.set_xlabels('ESTADO CIVIL')
cp.set_xticklabels(["OTRO", "CASADOS", "SOLTEROS", "DIVORCIADOS"])
```

```python
# CANTIDAD DE CLIENTES POR ESTADO CIVIL, CLASE DE CREDITO Y GRADO ACADEMICO
cp = sns.catplot('EDUCATION', col="IN_DEFAULT", data=credit, kind='count', hue='creditclass', order=[0,1,2,3,4,5,6], 
               hue_order=['class1','class2','class3','class4','class5','class6','class7','class8','class9','class10'],
                 aspect=1)
cp.fig.suptitle('CANTIDAD DE CLIENTES POR ESTADO CIVIL, CLASE DE CREDITO Y GRADO ACADEMICO',y=1.03)
cp.set_xlabels('GRADO ACADEMICO')
cp.set_xticklabels(["OTRO", "DIPLOMA", "UNIVERSIDAD", "BACHILLER", "OTRO", "OTRO", "OTRO"])
```
